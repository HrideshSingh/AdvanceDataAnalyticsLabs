# Importing

```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
df =  pd.read_csv('diamonds.csv')
sns.get_dataset_names()
```

```python
diamond=sns.load_dataset('diamonds', cache=True, data_home=None)
```

# 1. Scatter plot

```python
    sns.scatterplot(x="carat", y="price", hue="cut", size="depth", data=diamond)

    plt.xlabel('Carat Weight')

    plt.ylabel('Price')

    plt.title('Scatter Plot: Carat Vs Price (Based on Cut)')
```

# 2. Bar Plot

```python
    fig, ba_axis = plt.subplots(1, 3, figsize=(16,12))

    sns.barplot(x="cut", y="price", data=diamond, ax = ba_axis[0], color='Red')

    sns.barplot(x="color", y="price", data=diamond, ax = ba_axis[1], color='salmon')

    sns.barplot(x="clarity", y="price", data=diamond, ax = ba_axis[2], color='purple')
```

# 3. Line Plot


```python
    sns.lineplot(x="carat", y="x", data=diamond, color='salmon')
```

```python
sns.lineplot(x="carat", y="y", data=diamond, color='purple')
```

```python
sns.lineplot(x="carat", y="z", data=diamond, color='Red')
```

```python
sns.lineplot(x="carat", y="depth", data=diamond, sort=False, lw=1, color='blue')
```

# 4. Box Plot

```python
sns.boxplot(x="carat", y="cut", data=diamond, color="salmon", width=0.5, linewidth=2.5)
```

```python
sns.boxplot(x="carat", y="color", data=diamond, color="royalblue", width=0.5, linewidth=2.5)
```

```python
sns.boxplot(x="carat", y="clarity" , data=diamond, color="cyan", width=0.5, linewidth=2.5)
```

# 5. Pie Plot

```python
data = df.groupby("cut")["carat"].sum()
data.plot.pie(autopct="%.1f%%");
```
